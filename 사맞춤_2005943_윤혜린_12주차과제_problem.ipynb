3. (17 점) 아래 그림과 같은 버블 차트를 시각화하라. 

3.1 (1 점) 2015년 데이터만을 대상으로 아래 버블 차트를 시각화하라.  
3.2 (2 점) 차트 전체의 색상은 `blue`로, 투명도는 50%로 설정하라.  
3.3 (3 점) x 축 눈금 범위와 제목을 그림과 같이 설정하라.  
3.4 (3 점) y 축 눈금 범위와 제목을 그림과 같이 설정하라.  
3.5 (3 점) 버블 크기는 CCTV 비율(CCTV 대수를 인구 수로 나눈 값)로 설정하되, 축척 범위를 `[10, 1000]`으로 설정하고, 범례를 생략하라.  
3.6 (1 점) 작은 버블이 큰 버블보다 위에 그려지도록 하라.  
3.7 (2 점) 버블마다 툴팁으로 자치구 이름, CCTV 비율(소수점 4 자리수까지만 표시되도록 반올림 처리), 인구 수 및 CCTV 대수가 보이도록 하라.  
3.8 (2 점) 알테어가 제공하는 회귀 변환 기능을 적용하여, 그림과 같은 선형 회귀선을 표시하라.  

![1_bb_reg_2015](https://user-images.githubusercontent.com/10287629/169352281-81f93f63-9f2a-4929-b9ab-e45cf5e9a297.png)

```python
chart=alt.Chart(cctv_with_pop).transform_calculate(
    rate='datum.CCTV_규모_2015/datum.인구_규모_2015' #버블크기설정:transform_calculate 메소드
).mark_circle(color='blue',opacity=0.5 #차트 전체 스타일 지정
             
).encode(
    x=alt.X('인구_규모_2015:Q',
          scale=alt.Scale(type='linear', domain=[100000,700000]), #X축 설정
          title='인구 규모(단위:명)'),
    y=alt.Y('CCTV_규모_2015:Q',
          scale=alt.Scale(type='linear', domain=[0,7000]), #Y축 설정
          title='CCTV설치[단위:대]'),
    size=alt.Size('rate:Q', scale=alt.Scale(range=[10,1000]),legend=None), #설정한 버블 크기 적용
    order=alt.Order('rate:Q', sort='descending'), #작은 버블이 위로 가게
    tooltip=['자치구:N',alt.Tooltip('rate:Q',format='.4%'),'인구_규모_2015:Q','CCTV_규모_2015:Q']#툴팁
)

# chart + chart.transform_regression('인구_규모_2015:Q','datum.CCTV_규모_2015:Q').mark_line()

aa= alt.Chart(cctv_with_pop).transform_regression('인구_규모_2015','CCTV_규모_2015',method='linear'
).mark_line(color='blue',opacity=0.5
).encode(
    x=alt.X('인구_규모_2015:Q',
          scale=alt.Scale(type='linear', domain=[100000,700000]),
          title='인구 규모(단위:명)'),
    y=alt.Y('CCTV_규모_2015:Q',
          scale=alt.Scale(type='linear', domain=[0,7000]), #회귀 변환(선형)
          title='CCTV설치[단위:대]')
)

chart + aa # 최종 결과
```
